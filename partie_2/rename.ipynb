```python
import pandas as pd
import os

df_label = pd.read_csv("label_training_set.csv", sep = "\t", header = None)
df_label
# Chemin du dossier contenant les fichiers à renommer
folder_path = "training_set\\training_set\\"
# Initialise un compteur pour l'incrément
counter = 1

# Boucle sur les lignes de votre dataframe
for index, row in df_label.iterrows():
    if row[3] == "Car alarm":
        # Récupère le nom de fichier actuel et le nouveau nom de fichier
        current_file_name = str("Y")+ row[0]
        print(current_file_name)
        new_file_name = row[3].split(" ")[0]+ "_" + str(counter)
        # Construit le chemin complet vers le fichier actuel
        current_file_path = os.path.join(folder_path, current_file_name)
        # Construit le chemin complet vers le nouveau fichier
        new_file_path = os.path.join(folder_path, new_file_name)
        # Renomme le fichier
        os.rename(current_file_path, new_file_path)
        # Incrémente le compteur
        counter += 1
```
